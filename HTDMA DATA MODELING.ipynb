```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('all_day.txt', sep='\t')


scan = data[data['state'] == 'SCAN']
hold = data[data['state'] == 'HOLD']
flush = data[data['state'] == 'FLUSH']
done = data[data['state'] == 'DONE']


plt.loglog(scan['scanDiameter'], scan['electrometerN'], 
           marker='.', color='deepskyblue', linestyle='none', markersize=1, alpha=0.3, label='SCAN')

plt.loglog(hold['scanDiameter'], hold['electrometerN'], 
           marker='.', color='crimson', linestyle='none', markersize=1, alpha=0.2, label='HOLD')

plt.loglog(flush['scanDiameter'], flush['electrometerN'], 
           marker='.', color='limegreen', linestyle='none', markersize=1, alpha=0.2, label='FLUSH')

plt.loglog(done['scanDiameter'], done['electrometerN'], 
           marker='.', color='black', linestyle='none', markersize=1, alpha=0.2, label='DONE')


plt.xlabel('Scan Diameter')
plt.ylabel('Particle Concentration')
plt.title('Aerosol Size Distribution')


plt.legend(markerscale=10)

plt.show()
```
